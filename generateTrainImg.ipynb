```python
matFilePath = '/mnt/ibrixfs01-MRI/analysis/washen/temp/matFiles/'
outputPath = '/mnt/ibrixfs01-MRI/analysis/washen/temp/train/'
```

```python
import os
import glob
import scipy.io as sio
import numpy as np
import random
from skimage.transform import resize
```

```python
patients = glob.glob(matFilePath+'*T1*')
random.seed(40)
random.shuffle(patients)
nTrainPat = 118
sz = 256
```

```python
trainPatient = []
f = open("mean_sd.txt", "w")

for filename in patients[0:nTrainPat]:
    patName = str.split(filename,'/')[-1]
    patName = str.split(patName,'_')[0]
    print(patName)
    trainPatient.append(patName)
    
    #T1
    v = sio.loadmat(filename)
    v_t1 = v['V']
    m1 = np.mean(v_t1[v_t1>0])
    sd1 = np.std(v_t1[v_t1>0])
    
    #T2    
    v = sio.loadmat(matFilePath+patName+'_T2.mat')
    v_t2 = v['V']
    m2 = np.mean(v_t2[v_t2>0])
    sd2 = np.mean(v_t2[v_t2>0])
    
    f.write(patName+("\t%f\t%f\t%f\t%f")%(m1,sd1,m2,sd2) + "\n")
    
    #Segmentation
    v = sio.loadmat(matFilePath+patName+'-Segmentation.mat')
    bw = v['V']
    
    for i in range(v_t1.shape[2]):
        img = []
        t1 = resize(v_t1[:,:,i],(sz,sz),preserve_range=True)
        t2 = resize(v_t2[:,:,i],(sz,sz),preserve_range=True)
        img.append(t1)
        img.append(t2)

        seg = resize(bw[:,:,i],(sz,sz),preserve_range=True)
        
        if (np.sum(seg)>0):
            imgFilename = outputPath+'T1T2/'+patName+'-img-pos'+('_%03d.npy')%i
            segFilename = outputPath+'Segmentation/'+patName+'-seg-pos'+('_%03d.npy')%i
            np.save(imgFilename,img)
            np.save(segFilename,seg)
        else:
            imgFilename = outputPath+'T1T2/'+patName+'-img-neg'+('_%03d.npy')%i
            segFilename = outputPath+'Segmentation/'+patName+'-seg-neg'+('_%03d.npy')%i
            np.save(imgFilename,img)
            np.save(segFilename,seg)
    
```

```python
f.close()
```

```python
f = open("test_mean_sd.txt", "w")
testPath = '/mnt/ibrixfs01-MRI/analysis/washen/temp/test/'
for filename in patients[nTrainPat:-1]:
    patName = str.split(filename,'/')[-1]
    patName = str.split(patName,'_')[0]
    print(patName)
    
    #T1
    v = sio.loadmat(filename)
    v_t1 = v['V']
    m1 = np.mean(v_t1)
    sd1 = np.std(v_t1)
    
    #T2    
    v = sio.loadmat(matFilePath+patName+'_T2.mat')
    v_t2 = v['V']
    m2 = np.mean(v_t2)
    sd2 = np.mean(v_t2)
    
    f.write(patName+("\t%f\t%f\t%f\t%f")%(m1,sd1,m2,sd2) + "\n")
    
    #Segmentation
    v = sio.loadmat(matFilePath+patName+'-Segmentation.mat')
    bw = v['V']
    
    np.save(testPath+patName+"-T1.npy",v_t1)
    np.save(testPath+patName+"-T2.npy",v_t2)
    np.save(testPath+patName+"-BW.npy",bw)
```
